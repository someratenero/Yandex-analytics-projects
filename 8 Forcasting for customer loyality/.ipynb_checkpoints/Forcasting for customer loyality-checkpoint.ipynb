## Customer churn prediction model <a id="churn"></a>

[Back to content](#content)

Let's build a model of binary classification of clients where the target attribute is the fact of churn in the next month.

We will divide the data into features and the target variable "customer churn", and then divide the data into a training and validation sample.


```python
X = df.drop('Churn', axis=1) 
y = df['Churn'] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y) 
```

```python
for feature in X_train.columns:
    sns.distplot(df[feature], bins = 10)
    plt.show() 
```

Let's standardize the data.

```python
scaler = StandardScaler() 
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

```python
df_new = pd.DataFrame(data = X_train_st, columns = X_train.columns)
for feature in df_new.columns:
    sns.distplot(df[feature], bins = 10)
    plt.show() 
```

Let's train the model on the train sample using logistic regression.

```python
lr_model = LogisticRegression(random_state=0) 
lr_model.fit(X_train_st, y_train)
```

Let's use the already trained model to make predictions

```python
lr_predictions = lr_model.predict(X_test_st)
```

Now let's train the train sample using the model "Random forest". 

```python
rf_model = RandomForestClassifier (n_estimators=100, random_state=0)
rf_model.fit(X_train_st, y_train)
```

And make predictions. 

```python
rf_predictions = rf_model.predict(X_test_st) 
```

Let's evaluate the result of two models using the accuracy, precision, recall and f1 metrics on the validation sample. 

```python
print('Logistic regression metrics')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test,lr_predictions))) 
print('Precision: {:.2f}'.format(precision_score(y_test, lr_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, lr_predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, lr_predictions)))
print()
print('Random forest metrics')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test,rf_predictions))) 
print('Precision: {:.2f}'.format(precision_score(y_test, rf_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, rf_predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, rf_predictions)))
```

Summary:

- The proportion of correct answers in the training model "logical regression" and "random forest" is 90% (but this indicator is not quite suitable for us, because the classes are not balanced, the churn part is clearly less than a half),
- the accuracy of predictions is higher in the random forest: 79% for logical regression and 81% for random forest
- completeness is better in logical regression-82% compared to 75% in random forest.
- the balanced metric yields 81% in logical regression and 78% in random forest.

If it is important for us to catch all potentially outgoing clients, then it is better to focus on the completeness metric and logical regression coped with this task better. The summary metric f1 also indicates that the logical regression is more successful.

## Customers' clustering <a id="clusters"></a>

[Back to content](#content)

Let's standardize the table without a column with the churn value. 

```python
X_sc = scaler.fit_transform(X)
```

Let's create a distance matrix for standardized data using the linkage function

```python
linked = linkage(X_sc, method = 'ward') 
```

Here's the dendrogam with results: 

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.show() 
```

Bundles of four colors were formed, although the lilac one is so large that it can also be divided in halfs. Perhaps the orange branch too, so it would be optimal to set the maximum to 6 clusters but no more. Let's focus on 5 clusters.

Let's train a clustering model based on the K-Means algorithm and predict client clusters. 

```python
km = KMeans(n_clusters = 5)
labels = km.fit_predict(X_sc)
```

```python
df['labels']=labels
df.head()
```

```python
print('Silhouette_score: {:.2f}'.format(silhouette_score(X_sc, labels)))
```

The silhouette metric shows how much an object in one cluster is similar to another object in the same cluster compared to objects in another cluster. In our case, not very much. :)

Let's look at the average values of features for clusters.

```python
df.groupby(['labels']).mean()
```

Let's look at the cluster with the highest risk of outflow: they live closest than others, rarely use  partner programs or promotions, sign a contract for a minimum period, rarely go to group classes, they are the youngest of the clusters, spend the least, spent the least total time in the club and were less often than others in the club over the past month.

A cluster of the most loyal customers: they also live close, are about as infrequent in promo and partners programs but they sign much longer contracts, attend group classes more often, they are the oldest in the clusters, spend more money than all others, have been with the club for a long time and have come more often in the last month.

Let's build feature distributions for clusters. 

```python
plt.subplots(figsize=(20,15))
plots_in_row=len(df_num.columns.values)//num_cols + 1
for i, col in enumerate(df.columns.values):
    ax=plt.subplot(len(df.columns.values)//plots_in_row +1,plots_in_row, i+1)
    if col in ['Age','Avg_additional_charges_total', 'Lifetime',
       'Avg_class_frequency_current_month']:
        sns.boxplot(data=df, x='labels', y=col, ax=ax)
    else:
        sns.barplot(data=df,x='labels', y=col, ax=ax)
    plt.title(col)
plt.tight_layout()
plt.show()
```
